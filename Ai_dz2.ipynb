### Задание 1
У вас есть данные о книгах с полями: «Название», «Автор», «Год издания», «Цена».
Создайте DataFrame с не менее чем шестью строками.
Выведите весь DataFrame.
Определите среднюю цену книг.
Найдите книги, изданные после 2015 года.
Отсортируйте DataFrame по цене в порядке возрастания.

```python
import pandas as pd

books = {
    'Название': ['Война и мир', '1984', 'Ведьмак', 'Мастер и Маргарита', 'Empire V', 'Дюна', 'Transhumanism'],
    'Автор': ['Толстой', 'Оруэлл', 'Сапковский', 'Булгаков', 'Пелевин', 'Герберт', 'Пелевин'],
    'Год издания': [1869, 1949, 1992, 1967, 2006, 1965, 2021],
    'Цена': [1200, 600, 1500, 950, 800, 1000, 1100]
}
df = pd.DataFrame(books)
print(df)

print('Средняя цена:', df['Цена'].mean())

print('Книги после 2015:')
print(df[df['Год издания'] > 2015])

print('Сортировка по цене:')
print(df.sort_values('Цена'))
```

### Задание 2
В CSV-файле содержатся данные о заказах: «Номер заказа», «Клиент», «Дата», «Сумма».
Прочитайте данные из файла в DataFrame.
Выведите первые десять строк.
Определите количество заказов каждого клиента.
Найдите максимальную и минимальную суммы заказов.

```python
df_orders = pd.read_csv('orders.csv')
print(df_orders.head(10))

print(df_orders['Клиент'].value_counts())

print('Максимальная сумма:', df_orders['Сумма'].max())
print('Минимальная сумма:', df_orders['Сумма'].min())
print('Общая сумма:', df_orders['Сумма'].sum())
```

### Задание 3
Есть таблица с данными о продуктах питания: «Продукт», «Категория», «Калории», «Белки».
Создайте DataFrame с не менее чем десятью строками.
Выведите весь DataFrame.
Найдите все продукты с калорийностью выше 300.
Подсчитайте среднее количество белков по категориям.
Отсортируйте DataFrame по калорийности в порядке убывания.

```python
food_data = {
    'Продукт': ['Курица', 'Говядина', 'Яблоко', 'Банан', 'Творог', 'Сыр', 'Орехи', 'Рис', 'Макароны', 'Шоколад'],
    'Категория': ['Мясо', 'Мясо', 'Фрукты', 'Фрукты', 'Молочка', 'Молочка', 'Снеки', 'Крупы', 'Крупы', 'Сладости'],
    'Калории': [190, 250, 52, 89, 150, 350, 600, 130, 110, 540],
    'Белки': [27, 26, 0.3, 1.1, 18, 25, 15, 2.7, 3.5, 5]
}
df_food = pd.DataFrame(food_data)
print(df_food)

print('Калории > 300:')
print(df_food[df_food['Калории'] > 300])

print('Средние белки по категориям:')
print(df_food.groupby('Категория')['Белки'].mean())

print('Сортировка по калориям (убывание):')
print(df_food.sort_values('Калории', ascending=False))
```

### Задание 4
Собраны данные о сотрудниках и их проектах: «Имя», «Проект», «Часы».
Создайте DataFrame с не менее чем восемью строками.
Выведите исходный DataFrame.
Подсчитайте общее количество часов по каждому сотруднику.
Подсчитайте общее количество часов по каждому проекту.
Найдите сотрудника, затратившего больше всего часов.

```python
emp_data = {
    'Имя': ['Анна', 'Борис', 'Анна', 'Виктор', 'Борис', 'Дарья', 'Виктор', 'Дарья'],
    'Проект': ['Web', 'Mobile', 'Data', 'Web', 'Data', 'Mobile', 'Mobile', 'Web'],
    'Часы': [40, 35, 20, 45, 30, 25, 15, 38]
}
df_emp = pd.DataFrame(emp_data)
print(df_emp)

print('Часы по сотрудникам:')
print(df_emp.groupby('Имя')['Часы'].sum())

print('Часы по проектам:')
print(df_emp.groupby('Проект')['Часы'].sum())

print('Самый работящий:', df_emp.groupby('Имя')['Часы'].sum().idxmax())
```

### Задание 5
У вас есть таблица продаж билетов: «Фильм», «Город», «Проданные билеты».
Создайте DataFrame с не менее чем двенадцатью строками.
Выведите весь DataFrame.
Подсчитайте общее количество проданных билетов по каждому фильму.
Подсчитайте общее количество билетов по каждому городу.
Найдите фильм с наибольшим количеством продаж.

```python
tickets = {
    'Фильм': ['Дюна', 'Дюна', 'Дюна', 'Оппенгеймер', 'Оппенгеймер', 'Оппенгеймер', 'Барби', 'Барби', 'Барби', 'Бэтмен', 'Бэтмен', 'Бэтмен'],
    'Город': ['Баку', 'Москва', 'Лондон', 'Баку', 'Москва', 'Лондон', 'Баку', 'Москва', 'Лондон', 'Баку', 'Москва', 'Лондон'],
    'Проданные билеты': [150, 300, 450, 200, 250, 400, 180, 500, 350, 100, 150, 200]
}
df_t = pd.DataFrame(tickets)
print(df_t)

print('Билеты по фильмам:')
print(df_t.groupby('Фильм')['Проданные билеты'].sum())

print('Билеты по городам:')
print(df_t.groupby('Город')['Проданные билеты'].sum())

print('Самый популярный фильм:', df_t.groupby('Фильм')['Проданные билеты'].sum().idxmax())
```
