```python
import numpy as np

matr = np.random.randint(1, 51, size=(4, 4))
print("Матрица:")
print(matr)

print("Суммы по строкам:", np.sum(matr, axis=1))
print("Суммы по столбцам:", np.sum(matr, axis=0))
print("Общее среднее:", round(np.mean(matr), 2))
```

```python
massiv = np.arange(15).reshape(3, 5)
print("Исходный массив:")
print(massiv)

print("\nТранспонированный массив:")
print(massiv.T)

print("\nМассив после reshape:")
print(massiv.reshape(5, 3))
```

# Отчёт по продажам
*В этом разделе мы считаем итоги по простым данным.*

```python
sales = [120, 135, 100, 150, 90, 200, 170]

print("Сумма:", sum(sales))
print("Среднее:", round(sum(sales) / len(sales), 2))
print("Максимум:", max(sales))
print("Минимум:", min(sales))
```
