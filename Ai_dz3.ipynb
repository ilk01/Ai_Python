# Домашнее задание 3: Визуализация и Линейная регрессия

### Задание: Зависимость веса от роста
1. Создайте список ростов и список весов из файла `rost_ves.csv`.
2. Постройте точечный график (scatter plot), где по оси X — рост, а по оси Y — вес.
3. Добавьте подписи осей и заголовок «Зависимость веса от роста».
4. Сделайте точки зелёного цвета.
5. Обучите модель, чтобы она могла по росту предсказать вес.
6. Выведите метрики.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

df = pd.read_csv('rost_ves.csv')
X = df[['Рост (см)']].values
y = df['Вес (кг)'].values

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='green', label='Данные')
plt.xlabel('Рост (см)')
plt.ylabel('Вес (кг)')
plt.title('Зависимость веса от роста')
plt.grid(True)

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.plot(X, y_pred, color='red', linewidth=2, label='Линия регрессии')
plt.legend()
plt.show()

print(f'MSE: {mean_squared_error(y, y_pred):.2f}')
print(f'R2 Score: {r2_score(y, y_pred):.2f}')
print(f'Коэффициент наклона: {model.coef_[0]:.2f}')
```

### Задание 5: Температура в течение суток
1. Создайте список часов и список температур.
2. Постройте линейный график зависимости температуры от времени.
3. Добавьте заголовок «Температура в течение суток» и подписи осей.
4. Включите отображение сетки.
5. Обучите модель, чтобы она могла предсказывать температуру по часам.
6. Выведите метрики.

```python
hours = np.array([0, 4, 8, 12, 16, 20, 24]).reshape(-1, 1)
temp = np.array([15, 14, 18, 22, 25, 20, 17])

plt.figure(figsize=(10, 6))
plt.plot(hours, temp, marker='o', linestyle='-', color='blue', label='Реальная температура')
plt.title('Температура в течение суток')
plt.xlabel('Часы')
plt.ylabel('Температура (°C)')
plt.grid(True)

temp_model = LinearRegression()
temp_model.fit(hours, temp)
temp_pred = temp_model.predict(hours)

plt.plot(hours, temp_pred, color='orange', linestyle='--', label='Линия тренда (модель)')
plt.legend()
plt.show()

print(f'MSE для температуры: {mean_squared_error(temp, temp_pred):.2f}')
print(f'R2 Score для температуры: {r2_score(temp, temp_pred):.2f}')
```
